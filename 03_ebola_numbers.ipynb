```python
import pandas as pd
import numpy as np
import ast 
import datetime

file='ebola_2014_2016_clean.csv'
df_ebola = pd.read_csv(file)
```

```python
df_ebola
```

```python
df_ebola['year'] = df_ebola.apply(lambda row: int(row['Date'].split('-')[0]), axis=1)
df_ebola = df_ebola.rename(columns={'No. of confirmed cases': 'cases', 'No. of confirmed, probable and suspected deaths': 
                                   'deaths'})
df_ebola_yearly = df_ebola.groupby('year').sum()
```

```python
df_ebola_yearly.to_csv('ebola_yearly.csv')
```
